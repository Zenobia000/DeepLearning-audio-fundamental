# 音頻文件結構（Wave File Structure）

Wave 文件是一種標準的無損音頻文件格式，包含以下關鍵元素：

- **Header（文件頭）**：包括音頻格式、通道數、採樣率等基本信息。
- **Data（音頻數據）**：實際的聲波數據，由一連串的樣本組成。

## 音頻要素解釋

### Number of channels（通道數）
- 指單聲道（Mono）或立體聲（Stereo），即聲道的數量。單聲道的音頻只有一個聲道，而立體聲則有兩個聲道。

### Sample width（採樣寬度）
- 每個樣本的位元寬度，通常是 8 位或 16 位，決定了聲音的精度和動態範圍。

### Framerate/Sample Rate（取樣率）
- 每秒播放或錄製的樣本數，通常以 Hz 為單位（如 44,100 Hz）。取樣率越高，聲音細節越豐富。

### Number of frames（框數）
- 音頻數據中包含的總樣本數，反映了音頻的時長。

### Values of a frame（框的值）
- 每一框代表在某一瞬間的聲音振幅值，通常是數字表示聲音波形。


---
### 在數位音頻處理中，frame（框）是指音頻數據中的一個時間點的所有聲道數據的集合。



### 具體定義
每個 **frame** 包含一次取樣時所有通道（聲道）的取樣數據。

- **單聲道（Mono）** 音頻：每個 frame 只包含一個取樣值，因為只有一個聲道。
- **立體聲（Stereo）** 音頻：每個 frame 包含兩個取樣值，一個是左聲道的值，一個是右聲道的值。

### Frame 組成
- **一個 frame 的大小** 是由以下兩個因素決定的：
  - **Number of channels（通道數）**：frame 中包含的取樣數據的數量（如單聲道為 1，立體聲為 2）。
  - **Sample width（採樣寬度）**：每個取樣數據佔用的位元組數（如 16-bit 的採樣寬度需要 2 個位元組來表示一個取樣值）。

- 公式：
    ```text
    Frame size (bytes) = Number of channels × Sample width (bytes)
    ```


### Frame 與 Sample 的區別
- **Sample（樣本）**：每個通道在特定時間點上的取樣值，通常以 8-bit、16-bit 或 24-bit 等位元深度表示。對於單聲道音頻，每個 frame 就是一個 sample。
- **Frame（框）**：包含所有聲道的樣本。對於立體聲音頻，每個 frame 包含兩個樣本（左聲道和右聲道）。

#### 舉例
假設有一個 16-bit 立體聲（Stereo）音頻，則：
- **每個 sample** 為 16-bit（2 個位元組）。
- **每個 frame** 包含兩個 sample（左聲道和右聲道），所以 frame 的大小為 2（位元組） × 2（聲道） = 4 位元組。




## 利用 wave package 讀取音訊

````python
import wave
````

````python
# open wave file
audio_path = "./output.wav"
obj = wave.open(audio_path,'rb')

print("Number of channels", obj.getnchannels())
print("Sample width", obj.getsampwidth(), ",", 2*8, "-bit")
print("Framerate/Sample Rate.", obj.getframerate())
print("Number of frames", obj.getnframes())
print("parameters:", obj.getparams())
frames = obj.readframes(obj.getnframes())

print(len(frames) / obj.getsampwidth(), frames[0], type(frames[0]))
obj.close()

````

### Frame 的時間長度

**Frame 的時間長度**（即每個 frame 持續的時間）可以根據音頻的取樣率來計算。

公式：
```text
Frame duration (seconds) = 1 / Sample rate (Hz)
```

- **取樣率**（Sample rate）：每秒鐘的取樣數（通常以 Hz 為單位，如 44,100 Hz）。
- 每個 frame 的時間長度就是一個取樣的時間長度，表示一個 frame 持續的時間是多少秒。

#### 舉例
如果音頻的取樣率是 44,100 Hz，則每個 frame 的時間長度為：

Frame duration = 1 / 44100 ≈ 0.0000227 秒（22.7 微秒）

意味著每秒音頻包含 44,100 個 frame，且每個 frame 持續約 22.7 微秒。

````python
# 計算總時長
duration_seconds = obj.getnframes() / obj.getframerate()
print("Total duration (seconds):", duration_seconds)
````

## 利用 soundfile package 讀取音訊

````python
import soundfile as sf
````

````python
# 讀取音頻文件
audio_path = "./output.wav"
data, samplerate = sf.read(audio_path)


# 顯示音頻文件的相關信息
with sf.SoundFile(audio_path) as f:
    print('Number of channels:', f.channels)
    print('Subtype (Sample width):', f.subtype)  # 這裡的 subtype 表示音頻的樣本寬度格式 16bit
    print('Framerate/Sample Rate:', f.samplerate)
    print('Number of frames:', f.frames)
    print('Format:', f.format)

````

````python
# 儲存音頻文件
output_path = "new_file_soundfile.wav"  # 要儲存的文件路徑
sf.write(output_path, data, samplerate, subtype=f.subtype)  # 使用原有樣本寬度重新保存
````

## 利用 librosa package 讀取音訊

````python
import librosa
import soundfile as sf
import numpy as np

# 讀取音頻數據
audio_path = "./output.wav"
data, sample_rate = librosa.load(audio_path, sr=16000)  # 將取樣率設置為 16000 Hz

# 顯示音頻文件的相關信息
print("Number of channels:", 1 if data.ndim == 1 else data.shape[0])
print("Framerate/Sample Rate:", samplerate)
print("Number of frames:", data.shape[0])
print("Data type:", data.dtype)
````

````python
# 將浮點數據轉換為 16-bit PCM 格式
data_int16 = np.int16(data * 32767)  # 將 [-1.0, 1.0] 的 float32 轉換為 int16

# 寫入音頻文件
output_path = "new_file_librosa.wav"
sf.write(output_path, data_int16, sample_rate, subtype='PCM_16')  # 16-bit 單聲道
````
