## 傅立葉級數 (Fourier Series) 與 傅立葉積分 (Fourier Integral) 的適用場合
### 傅立葉級數 (Fourier Series)
傅立葉級數主要適用於週期性的信號或函數。它將一個週期函數展開成一組正弦函數和餘弦函數的無限級數之和。這些正弦和餘弦函數的頻率是基於該週期函數的基本頻率及其整數倍。這種表示方法適合描述在有限範圍內的振盪或波動現象。

> 適用場合：
- 週期性函數，如正弦波、方波等。
    - 電路分析中的交流信號。
    - 任何具有固定重複模式的物理系統，如聲波、電磁波等。

### 傅立葉積分 (Fourier Integral)
當函數是非週期性時，傅立葉級數無法直接應用，這時可以使用傅立葉積分來表示。傅立葉積分將一個非週期函數分解為無窮多正弦波和餘弦波的疊加，這些波的頻率可以是連續的。傅立葉積分是傅立葉變換的基礎，它能將時間域的信號轉換到頻率域中。

> 適用場合：
- 非週期性函數，如脈衝信號、衰減波形等。
    - 信號處理中的頻譜分析。
    - 非穩態熱傳導分析。
    - 任何描述非週期性現象的場合，如無限長的電磁波傳播、光學系統的衍射分析等。
